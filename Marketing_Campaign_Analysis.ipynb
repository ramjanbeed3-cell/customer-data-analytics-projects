```python
# 📊 Marketing Campaign Effectiveness Analysis
# Author: [Your Name]
# Objective: Analyze campaign data to identify top-performing marketing channels based on ROI and conversions.

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

print("✅ Libraries imported successfully")

# Step 2: Create or Load Dataset
data = {
    'Campaign': ['Email', 'Social Media', 'Google Ads', 'Affiliate', 'TV', 'Radio', 'Influencer'],
    'Channel': ['Email', 'Social', 'Search', 'Referral', 'Offline', 'Offline', 'Social'],
    'Spend($)': [5000, 7000, 9000, 3000, 12000, 4000, 6000],
    'Leads': [400, 600, 1000, 200, 1500, 300, 800],
    'Conversions': [80, 90, 150, 50, 160, 40, 120],
    'Revenue($)': [12000, 13500, 23000, 7000, 27000, 6000, 18000],
    'Region': ['North', 'South', 'East', 'West', 'North', 'East', 'South']
}

df = pd.DataFrame(data)
print("\n✅ Sample Marketing Campaign Data Created:")
display(df)

# Step 3: Data Cleaning
print("\n🧹 Checking for missing values:")
print(df.isnull().sum())

print("\n🧾 Data Types:")
print(df.dtypes)

# Convert Revenue and Spend columns to numeric (in case of any non-numeric values)
df['Spend($)'] = pd.to_numeric(df['Spend($)'], errors='coerce')
df['Revenue($)'] = pd.to_numeric(df['Revenue($)'], errors='coerce')

# Step 4: Descriptive Analytics (KPIs)
df['Conversion Rate (%)'] = (df['Conversions'] / df['Leads']) * 100
df['ROI (%)'] = ((df['Revenue($)'] - df['Spend($)']) / df['Spend($)']) * 100
df['Revenue per Lead ($)'] = df['Revenue($)'] / df['Leads']

print("\n📈 Campaign Performance Metrics:")
display(df)

# Overall Summary
total_spend = df['Spend($)'].sum()
total_revenue = df['Revenue($)'].sum()
overall_roi = ((total_revenue - total_spend) / total_spend) * 100

print(f"\n💰 Total Spend: ${total_spend}")
print(f"💵 Total Revenue: ${total_revenue}")
print(f"📊 Overall ROI: {overall_roi:.2f}%")

# Step 5: Visualization - ROI by Campaign
plt.figure()
sns.barplot(x='Campaign', y='ROI (%)', data=df, palette='viridis')
plt.title("ROI by Marketing Campaign")
plt.xticks(rotation=45)
plt.show()

# Step 6: Visualization - Leads vs Conversions
plt.figure()
sns.scatterplot(x='Leads', y='Conversions', hue='Campaign', data=df, s=120)
plt.title("Leads vs Conversions by Campaign")
plt.show()

# Step 7: Visualization - Spend vs Revenue
plt.figure()
sns.barplot(x='Campaign', y='Revenue($)', data=df, color='green', label='Revenue')
sns.barplot(x='Campaign', y='Spend($)', data=df, color='red', label='Spend')
plt.title("Campaign Spend vs Revenue Comparison")
plt.legend()
plt.show()

# Step 8: Advanced Analytics - Correlation
print("\n📊 Correlation Matrix (numerical features):")
corr = df[['Spend($)', 'Leads', 'Conversions', 'Revenue($)', 'ROI (%)']].corr()
display(corr)

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()

# Step 9: Insights Summary
top_roi = df.loc[df['ROI (%)'].idxmax()]
top_conversion = df.loc[df['Conversion Rate (%)'].idxmax()]
top_revenue = df.loc[df['Revenue($)'].idxmax()]

print("\n🔍 Key Insights:")
print(f"🔥 Highest ROI Campaign: {top_roi['Campaign']} ({top_roi['ROI (%)']:.2f}%)")
print(f"🏆 Best Conversion Rate: {top_conversion['Campaign']} ({top_conversion['Conversion Rate (%)']:.2f}%)")
print(f"💸 Highest Revenue Campaign: {top_revenue['Campaign']} (${top_revenue['Revenue($)']})")
print(f"\n📊 Average ROI across all campaigns: {df['ROI (%)'].mean():.2f}%")

# Step 10: Save Results
df.to_csv("marketing_campaign_summary.csv", index=False)
print("\n✅ Analysis complete! Results saved to 'marketing_campaign_summary.csv'")
```
